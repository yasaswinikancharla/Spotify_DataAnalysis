```python
spotify_hist = spotify_features.drop(spotify_features.columns[[0, 1, 2, 3, 4, 10, 16]], axis=1)

# Plot histograms for all numerical columns
spotify_hist.hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# Calculate average popularity by genre
genre_popularity = (
    spotify_features.groupby('genre')['popularity']
    .mean()
    .round()
    .reset_index()
    .rename(columns={'popularity': 'average_popularity'})
)

# Sort genres by average popularity
genre_popularity = genre_popularity.sort_values(by='average_popularity', ascending=False)

# Create the plot
plt.figure(figsize=(10, 8))
sns.barplot(
    data=genre_popularity,
    y='genre',  # Horizontal bar plot
    x='average_popularity',
    palette='viridis'
)

# Customize the plot
plt.title('Average Popularity by Genre', fontsize=16)
plt.xlabel('Average Popularity', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
popular = spotify_features.sort_values('popularity', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = popular).set(title='Top 5 Genres by Popularity')
```

```python
spotify_green = '#1DB954' 
# Set the background color
plt.figure(facecolor='black')
sns.histplot(data=spotify_tracks, x='duration',bins=150, color=spotify_green) 
plt.xlabel('Track Duration (Seconds)', color='white')
plt.ylabel('Number of Tracks', color='white')
plt.title('Distribution of Track Durations', color='white')
plt.xticks(color='white') 
plt.yticks(color='white') 
plt.show()
```

```python
#Artist with highest number of tracks
top_artist =spotify_tracks['artists'].value_counts().idxmax()
print(f"Artist with the most tracks: {top_artist}")
```

```python
years=spotify_tracks.release_date.dt.year
```

```python
total_dr = spotify_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```

```python
sns.displot(years, discrete=True, aspect=2, height=4, kind='hist',color='g').set(title='No of songs - per year')
```

```python
#Sample 8 percent of total dataset.
sample_sp=spotify_tracks.sample(int(0.008*len(spotify_tracks)))
print(len(sample_sp))
```

```python
#Create a Regression Plot Between Loudness and Energy. Let’s plot it  in the form of a Regression Line.
plt.figure(figsize=(8,4))
sns.regplot(data=sample_sp, y='loudness', x='energy', color='#054907').set(title='Regression Plot - Loudness vs Energy Correlation')
```

```python
#Create a Regression Plot Between Popularity and Acousticness in the Form of a Regression Line.
plt.figure(figsize=(8,4))
sns.regplot(data=sample_sp, y='popularity', x='acousticness', color='#008000').set(title='Regression Plot - Popularity vs Acousticness Correlation')
```
